```python
import matplotlib.pyplot as plt
import numpy as np
import SimpleITK as sitk
import glob
import pandas as pd
import os
from sklearn.model_selection import train_test_split
from Features import *  

def cut_image(img_names,mask_names):
    mask=plt.imread(mask_names)
    img=sitk.ReadImage(img_names)
    img_array=sitk.GetArrayFromImage(img)[0]
    arrays = (mask * img_array)
    sums =np.sum(arrays,axis=1)
    for i in range(len(sums)):
        if sums[i]!=0:
            arrays=arrays[i-20:i+80,200:300]
            break
    return arrays.reshape(100,100,1)

def files():
    files = pd.read_csv('clinical_data.csv', sep=',', encoding='gbk')
    files = files.iloc[-10:]
    values = files.values
    cnndatas=[]
    featuredatas=[]
    labels=[]
    j=0
    for val in values:
        name = val[2]
        label = val[3]
        if val[0]=='m':
            sex=1
        else:
            sex=0
        age=int(val[1][:2])
        imgnames = glob.glob('./dataset/' + str(name) + '/' + 'arterial phase' + '/*.dcm')
        masknames = glob.glob('./dataset/' + str(name) + '/' + 'arterial phase' + '/*.png')
        for img, mask in zip(imgnames, masknames):
            array = plt.imread(mask)
            sums = np.sum(array)
            if sums != 0:
                featur, lens = features(mask, img)
                featur.append(sex)
                featur.append(age)
                cnndatas.append(cut_image(img, mask))
                featuredatas.append(featur)
                if label == '+':
                    labels.append(1)
                else:
                    labels.append(0)
                j += 1
                if j % 10 == 0:
                    print("数据加载：", j)
    cnndatas=np.array(cnndatas)
    print('cnndatas:',cnndatas)
    featuredatas=np.array(featuredatas)
    print('featuredatas:',featuredatas)
    labels=np.array(labels)
    print('labels:',labels)
    print("保存数据集")
    np.save('cnndata.npy', cnndatas)
    np.save('label.npy', labels)
    np.save('featuredata.npy', featuredatas)
    return  cnndatas,featuredatas,labels
cnndatas,featuredatas,labels = files()
```

## 2 模型训练

```python
import tensorflow as tf

cnndata=np.load('cnndata.npy',allow_pickle=True)
featuredata=np.load('featuredata.npy',allow_pickle=True)
label=np.load('label.npy',allow_pickle=True)

train_cnn, test_cnn, train_label, test_label = train_test_split(cnndata, label, test_size=0.2,random_state=1)

train_feature,test_feature,feature_train_label,feature_test_label=train_test_split(featuredata, label, test_size=0.2,random_state=1)

num_classes=len(set(train_label))


datas_placeholder = tf.placeholder(tf.float32, [None, 100,100,1])
labels_placeholder = tf.placeholder(tf.int32, [None])
dropout_placeholdr = tf.placeholder(tf.float32)
def average_pool():
    conv0 = tf.layers.conv2d(datas_placeholder, 20, 5, activation=tf.nn.relu)
    pool0 = tf.layers.average_pooling2d(conv0, [2, 2], [2, 2])
    conv1 = tf.layers.conv2d(pool0, 40, 4, activation=tf.nn.relu)
    pool1 = tf.layers.average_pooling2d(conv1,  [2, 2], [2, 2])
    conv2 = tf.layers.conv2d(pool1, 80, 2, activation=tf.nn.relu)
    pool2 = tf.layers.average_pooling2d(conv2,  [2, 2], [2, 2])
    return pool2

pool2=average_pool()
flatten = tf.layers.flatten(pool2)
fc = tf.layers.dense(flatten, 450 , activation=tf.nn.relu)
dropout_fc = tf.layers.dropout(fc, dropout_placeholdr)
logits = tf.layers.dense(fc,num_classes)
predicted_labels = tf.argmax(logits, 1)
losses = tf.nn.softmax_cross_entropy_with_logits(
    labels=tf.one_hot(labels_placeholder, num_classes),
    logits=logits)


mean_loss = tf.reduce_mean(losses)
optimizer = tf.train.AdamOptimizer(learning_rate=1e-2).minimize(losses)
saver = tf.train.Saver()
def run_trains(model_path):
    with tf.Session() as sess:
        print("训练模式")
        sess.run(tf.global_variables_initializer())
        batch_size=60
        input_count=len(train_cnn)
        batches_count=int(input_count/batch_size)
        remainder=input_count%batch_size
        print("数据集分成 %s 批, 前面每批 %s 个数据，最后一批 %s 个数据" % (batches_count + 1, batch_size, remainder))
        for step in range(60):
            for n in range(batches_count):
                train_feed_dict = {
                    datas_placeholder: train_cnn[n*batch_size:(n+1)*batch_size],
                    labels_placeholder: train_label[n*batch_size:(n+1)*batch_size],
                    dropout_placeholdr: 0.25,
                }
                _, mean_loss_val = sess.run([optimizer, mean_loss], feed_dict=train_feed_dict)
                print("n = {}%tmean loss = {}".format(n, mean_loss_val))
            if remainder>0:
                start_index = batches_count * batch_size
                train_feed_dict = {
                    datas_placeholder: train_cnn[start_index:input_count-1],
                    labels_placeholder: train_label[start_index:input_count-1],
                    dropout_placeholdr: 0.25,
                }
            _, mean_loss_val = sess.run([optimizer, mean_loss], feed_dict=train_feed_dict)
            print("step = {}%tmean loss = {}".format(step, mean_loss_val))
            print("保存模型")
            saver.save(sess, model_path)
            print("训练结束，保存模型到{}".format(model_path))
def run_tests(model_path):
    with tf.Session() as sess:
        print("测试模式")
        saver.restore(sess, model_path)
        print("从{}载入模型".format(model_path))
        test_feed_dict = {
            datas_placeholder: test_cnn,
            labels_placeholder: test_label,
            dropout_placeholdr: 0
        }
        predicted_labels_val = sess.run(predicted_labels, feed_dict=test_feed_dict)
        m=0
        j=0
        for real_label, predicted_label in zip(test_label, predicted_labels_val):
            if real_label==predicted_label:
                m+=1
            j+=1
        print("预测准确率：{}".format(float(m/j)))
    return predicted_labels_val
def get_features(model_path,x_data,y_data):
    with tf.Session() as sess:
        saver.restore(sess, model_path)
        test_feed_dict = {
            datas_placeholder: x_data,
            labels_placeholder: y_data,
            dropout_placeholdr: 0
        }
        predicted_labels_val = sess.run(fc, feed_dict=test_feed_dict)
    return predicted_labels_val

def train_feature(traindata,trainlabel,testdata,testlable,model_path,i):
    clf = RandomForestClassifier(n_estimators=i,max_depth=None,random_state=0)
    clf.fit(traindata,trainlabel)
    joblib.dump(clf,paths)

def make_randomforest(testdata,testlabel):
    clf = joblib.load(model_path)
    values = cf.predict(testdata)
    return values

run_trains('./dataset')
```
