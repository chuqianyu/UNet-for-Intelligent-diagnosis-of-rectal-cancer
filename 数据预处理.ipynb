```python
# 导入库
import SimpleITK as sitk
import numpy as np
import os
import re
import glob
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split

model_path = './dataset'

def filename(model_path):
    names = []
    imglist = []
    labels = []
    for name in os.listdir(model_path):
        pattern = re.compile('\d{4}')
        if pattern.findall(name) != []:
            names.append(name)
        for i in names:
            imgs = glob.glob(model_path + '/' + i +'/' +'arterial phase'+'/*.dcm')
            masks = glob.glob(model_path + '/' + i +'/' +'arterial phase'+'/*.png')
            for mask,img in zip(masks,imgs):
                array = plt.imread(mask)
                sums = np.sum(array)
                imglist.append(img)
                if sums != 0:
                    labels.append(1)
                else:
                    labels.append(0)
    return labels,imglist,names
labels,imglist,names = filename(model_path)
```

## 2 数据归约

```python
datas = []
for data in imglist:
    img = sitk.ReadImage(data)
    img_array = sitk.GetArrayFromImage(img)
    img_array = img_array[0,230:430,180:320] 
    datas.append(np.array(img_array).reshape(200,140,1))

datas = np.array(datas)
labels = np.array(labels)

train_datas,test_datas,train_labels,test_labels = train_test_split(
    datas,labels,test_size = 0.2,random_state = 123)

print('训练集大小',len(train_datas))
print('测试集大小',len(test_datas))

np.save('./tmp/train_datas.npy',train_datas)
np.save('./tmp/test_datas.npy',test_datas)
np.save('./tmp/train_labels.npy',train_labels)
np.save('./tmp/test_labels.npy',test_labels)
```
